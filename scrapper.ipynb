```python
import requests
import bs4
```

```python
#Getting Data Requests
request1=requests.get('https://www.flipkart.com/motorola-edge-50-pro-5g-68w-charger-luxe-lavender-256-gb/p/itmba8105d0d22ad?pid=MOBGXFXYMTKAMPTS&lid=LSTMOBGXFXYMTKAMPTSJQVWQM&marketplace=FLIPKART&store=tyy%2F4io&spotlightTagId=BestsellerId_tyy%2F4io&srno=b_1_1&otracker=browse&fm=organic&iid=2e473742-64dc-4b47-a253-87c18953fe6f.MOBGXFXYMTKAMPTS.SEARCH&ppt=browse&ppn=browse&ssid=g717zxebphjfjs3k1715079559978')
request1
```

```python
request1.content
```

```python
soup=bs4.BeautifulSoup(request1.text)
soup
```

```python
#Fetching Real time Comments and Reviews:
reviews=soup.find_all('div',{'class':"ZmyHeo"});
for review in reviews:
    print(review.get_text()+'\n\n')
```

```python
#individual ratings:
indi_ratings=soup.find_all('div',{'class':'XQDdHH Ga3i8K'});
for individual_rating in indi_ratings:
    print(individual_rating.get_text() + '\n')
```

```python
#comment tags:
tags=soup.find('span',{'class':'_6EBuvT OHdLLG'}).get_text();
tags
```

```python
#customer name:
customer_name=soup.find_all('p',{'class':'_2NsDsF AwS1CA'});
for cust_name in customer_name:
    print(cust_name.get_text() +'\n')
```

```python
#Average ratings:
ratings=soup.find('div',{'class':'ipqd2A'}).get_text()
print(ratings)
```

```python
#Price of product :
product_price=soup.find('div',{'class':'Nx9bqj CxhGGd'}).get_text();
print(product_price)
```
