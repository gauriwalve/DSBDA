```python
#Perform the following operations using Python on the Facebook metrics data sets 
#a. Create data subsets using loc() method 
#b. Merge Data 
#c. Sort Data (Descending Order) 
#d. Transposing Data 
#e. Reshape Data 
```

```python
#importing the library
import pandas as pd
```

```python
#reading the dataset
df = pd.read_csv('dataset_Facebook.csv',delimiter=';')
```

```python
df
```

```python
#a. Create data subsets using loc() method 
# a.1 Extract data of specific rows of a dataframe:
df.loc[[0,1,3]]
```

```python
# a.2 Create a subset of rows using slicing:
df.loc[0:5:10]
```

```python
# a.3 Create a subset of particular columns using Labels:
df_subset_1= df.loc[0:4,['like', 'share']]
df_subset_1
```

```python
df_subset_2= df.loc[0:4,['like', 'Category']]
df_subset_2
```

```python
#b.Merge data
merged_data = pd.merge(df_subset_1, df_subset_2, on='like')
merged_data
```

```python
#another method
pd.concat([df_subset_1, df_subset_2])
```

```python
#c. Sort Data (Descending Order)
# Sorting data descending order wrt 'Like'
df.sort_values(by=['like'],ascending=False)
```

```python
#d. Transposing Data
# Method 1
merged_data.transpose()
```

```python
# Method 2
merged_data.T
```

```python
#e. Reshape Data 
# Comment is id_vars and Type is value_vars
pd.melt(df, id_vars =['Type'], value_vars =['comment'])
```
